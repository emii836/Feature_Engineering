```python
def count_common(x):
    a=np.array(x)
    b=np.where(a==np.min(a))
    v=np.array(b)
    nr_ap=v.shape[1]
    return nr_ap
```

```python
def count_least(x):
    a=np.array(x)
    b=np.where(a==np.max(a))
    v=np.array(b)
    nr_ap=v.shape[1]
    return nr_ap
```

```python
X["COMMON_COUNT_FEAT"]=X.apply(count_common,axis=1)
X["LEAST_COUNT_FEAT"]=X.apply(count_least,axis=1)
```
