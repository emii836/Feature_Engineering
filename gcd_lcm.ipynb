```python
def  kin_energy(x):
    f=np.unique(x,return_counts=True)
    prob=f[1]/x.shape[0]
    energy=np.sum(prob**2)
    return  energy
```

```python
def count_common(x):
    a=np.array(x)
    b=np.where(a==np.min(a))
    v=np.array(b)
    nr_ap=v.shape[1]
    return nr_ap
```

```python
def count_least(x):
    a=np.array(x)
    b=np.where(a==np.max(a))
    v=np.array(b)
    nr_ap=v.shape[1]
    return nr_ap
```

```python
X["KINETIC_FEAT"]=X.apply(kin_energy,axis=1)

X["max_FEAT"]=X[good_cols].apply(np.max,axis=1)

X["mean_FEAT"]=X[good_cols].apply(np.mean,axis=1)
X["COMMON_COUNT_FEAT"]=X.apply(count_common,axis=1)
X["LEAST_COUNT_FEAT"]=X.apply(count_least,axis=1)
```

```python
def gcd(a, b):
    if a == 0 or b == 0:
        return max(a, b)
    return gcd(b, a % b)
```

```python
X["GCD_FEAT"]=X.apply(lambda r: gcd(r["feature_elusive_vapoury_accomplice"],r["feature_fortified_repressed_baird"]),axis=1)

```

```python
def l(a, b):
    if a==0 or b==0:
        return max(a, b)
    else:
        if a!=0 and b!=0:
           lcm=(a*b)/gcd(a, b)
    if a==0 and b==0:
        return 0
    return lcm
```
